**for wine quality we need to check if it is greter than the 7 or not.**</br>
**if it is grater than 7 os it is good wine and less then the 7 it is not an good wine.**

```python
df['quality'].value_counts()
```

```python
df['good quality']=df['quality'].apply(lambda x: 1 if x>=7  else 0)
```

```python
df.head()
```

```python
df.drop(['quality'],axis=1,inplace=True)
```

```python
df.head()
```

## Model selection

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

**Logiastic Regresstion**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor=LogisticRegression()
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```
